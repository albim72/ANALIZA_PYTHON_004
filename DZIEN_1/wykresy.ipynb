```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
osoby = ("Anna","Piotr","Olga","Roman")
informacje={
    'wiek':(34,56,23,45),
    'waga':(56.8,89.6,60.3,103.6),
    'wzrost':(162,172,170,189)
}
```

```python
x = np.arange(len(osoby))
width = 0.25
multiplier = 0
```

```python
fig,ax = plt.subplots(constrained_layout=True)
for attribute, mesurement in informacje.items():
    offset = width*multiplier
    rects = ax.bar(x+offset,mesurement,width,label = attribute)
    ax.bar_label(rects,padding=5)
    multiplier += 1

ax.set_ylabel("cecha")
ax.set_title("wybrane cechy osób...")
ax.set_xticks(x+width,osoby)
ax.legend(loc='upper left')
ax.set_ylim(0,220)

plt.show()
```

```python
fig,ax = plt.subplots()
marki = ['audi','toyota','mercedes','honda','kia','jeep','dodge']
sprzed = [12.7,28.9,9.8,16.8,23.5,7.8,12.1]

bar_labels = ['red','green','_red','blue','orange','_red','_green']
bar_colors = ['tab:red','tab:green','tab:red','tab:blue','tab:orange','tab:red','tab:green']

ax.bar(marki,sprzed,label=bar_labels, color=bar_colors)
ax.set_xlabel('sprzedaż wybranych marek')
ax.set_ylabel('marki samochodów')
ax.legend(title='marki')
plt.show()
```

```python
cases = [
    None,
    8,
    (30,8),
    [16,24,32],
    [0,-1],
    slice(100,200,3),
    0.1,
    0.4,
    (0.2,0.4)
]
delta = 0.11
x = np.linspace(0,10-2*delta,200)+delta
y = np.sin(x) +1.0 + delta
```

```python
fig, axs = plt.subplots(3,3,figsize=(10,6), layout='constrained')
for ax, markevery in zip(axs.flat,cases):
    ax.set_title(f'markevery = {markevery}')
    ax.plot(x,y,'o',ls='-',ms =4,markevery=markevery)

plt.show()
```
